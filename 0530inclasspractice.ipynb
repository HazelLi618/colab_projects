```python
import nltk
from nltk.corpus import PlaintextCorpusReader
corpus_root = "/content/drive/MyDrive/lanalyrics"
wordlists = PlaintextCorpusReader(corpus_root,'.*')
songs = wordlists.fileids()
print(songs)
```

```python
import nltk
testText = wordlists.words("born to die lyrics.txt")
single = nltk.text.Text(testText)
findList = ["a","the","can","will"]
single.dispersion_plot(findList)
```

```python
songs = wordlists.fileids()
cfd = nltk.ConditionalFreqDist((fileid,word)
    for fileid in songs
    for word in wordlists.words(fileid))
findList = ["a","the","can","will","should","must"]
cfd.tabulate(conditions=songs,samples=findList)
```

```python
cfd = nltk.ConditionalFreqDist(
    (target,fileid[:10])
    for fileid in wordlists.fileids()
    for w in wordlists.words(fileid)
    for target in ["baby","love"]
    if w.lower() == target)
cfd.plot()
```

```python
import matplotlib.pyplot as plt
x = ["love","baby","bae","you"]
y = [0,5,10,15]

plt.bar(x,y)
plt.title("key words")
plt.xlabel("words")
plt.ylabel("nums")
plt.figure(figsize=(5,4))
plt.show()
```
